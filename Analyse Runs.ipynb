```python
# See http://geophysicslabs.com/2016/06/18/basic-statistics-with-parkrun-data/
#     https://github.com/jobar8/parkrun
import parkrun
import matplotlib.pyplot as plt


inputs = {
    'Hampstead Heath': 'hh-2019-05-11-tabs.txt',
    'Royal Tunbridge Wells': 'tw-2019-06-01-tabs.txt',
}

def get_runs():
    for name, infile in inputs.items():
        yield name, parkrun.importResults(
                infile, skiprows=12, skipfooter=38
            )
    
runs = dict(get_runs())


for name, run in runs.items():
    print(name)
    parkrun.print_stats(run)
    print()
```

```python
run.head()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1)

for i, (name, run) in enumerate(runs.items()):
    print(name)
    run['Time'].plot.bar(
        figsize=(20, 10), 
        color=[('r' if r['Runner']=='Tom VINER' else 'b') for i, r in run.iterrows()],
        ax=axes[i]
    )
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1)

for i, (name, run) in enumerate(runs.items()):
    print(name)
    age_sort_run = run.sort_values(by='Age Grade', ascending=False)
    age_sort_run['Age Grade'].plot.bar(
        figsize=(20, 10), 
        color=[('r' if r['Runner']=='Tom VINER' else 'b') for i, r in age_sort_run.iterrows()],
        ax=axes[i]
    )
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1)

for i, (name, run) in enumerate(runs.items()):
    print(name)
    age_sort_run = run.sort_values(by='Total Runs', ascending=False)
    age_sort_run['Total Runs'].plot.bar(
        figsize=(20, 10), 
        color=[('r' if r['Runner']=='Tom VINER' else 'b') for i, r in age_sort_run.iterrows()],
        ax=axes[i]
    )
```
